```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
file_path = "drive/My Drive/final_hate_with_google.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
df.head(3)
```

```python
normalizer = Normalizer()
```

```python
def normal_text(row):
  text = str(row.text)
  return normalizer.normalize(text)
```

```python
df['norm_text'] = df.apply(normal_text , axis = 1)
```

```python
df.head()
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
tokenizer.src_lang = "fa_IR"
cuda0 = torch.device('cuda:0')
model.to('cuda:0')
```

```python
def translate_using_mbart(row):
  text = str(row.norm_text)
  encoded_fa = tokenizer(text, return_tensors="pt") 
  input_ids = torch.tensor(encoded_fa['input_ids'].tolist(), dtype=torch.long, device=cuda0)
  attention_mask = torch.tensor(encoded_fa['attention_mask'].tolist(), dtype=torch.long, device=cuda0)
  generated_tokens = model.generate(input_ids=input_ids, attention_mask=attention_mask, forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"])
  print('x')
  return tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
res = []
def encode_tofa(row):
  text = str(row.norm_text)
  encoded_fa = tokenizer(text, return_tensors="pt")
  res.append(encoded_fa)
  return encoded_fa
```

```python
df['mbart'] = df.apply(translate_using_mbart, axis = 1)
```

```python
df['mbart'] = df.apply(lambda x: x.mbart[0], axis = 1)
```
